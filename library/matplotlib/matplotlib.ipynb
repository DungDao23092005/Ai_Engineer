### Pyplot APi vs Object-Oriented (OO) API
* pyplot API -> Quickly
* OO API -> advanced

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color = 'Blue',linestyle='dashed',label='sin(x)');
plt.plot(x,np.cos(x),color = 'Red', label ="cos(x)")

plt.title("Sin(x) and Cos(x)")
plt.xlabel("Bien X")
plt.ylabel("Sin(X)")
#plt.xlim([0,4])
#plt.ylim([0,4])
#plt.axis([0,4,-0.7,0.6])
plt.axis("equal")

plt.legend();
```

# Object-Oriented API

```python
#1. Prepare data
x=[1,2,3,4]
y= [11,22,33,44]

#2. Setup plot
fig , ax = plt.subplots(figsize=(5,5)) # wight and height

#3. plot the data
ax.plot(x,y)

ax.set(title="A simple plot ",
           xlabel ="x-axis",
           ylabel="y-axis");
```

# 2.Most common type of Matplotlib plots các loại đồ thị trong Matplotlib
* line
* scatter
* bar
* hist
* subplots()

```python
plt.plot();
```

```python
# Create an array
x=np.linspace(0,10,100)
x[:5]
```

```python
# OO API plot a line chart

fig,ax = plt.subplots()
ax.plot(x,x**3)
```

```python
#Pylot API
plt.scatter(x,np.exp(x)); #y=e^(x)
```

```python
# OO API
#Prepare new data
rng=np.random.RandomState(0)

x=rng.randn(100)
y=rng.randn(100)
color=rng.rand(100)
sizes = 1000*rng.rand(100)

fig,ax = plt.subplots()
img1=ax.scatter(x,y, s=sizes,c=color,cmap='viridis',alpha=0.3);

fig.colorbar(img1)
```

### 3.Bar
* Vertical(trục dọc)
* Horizontal(trục ngang)

```python
soft_drink_prices ={
    "Coke":10,
    "Pessi":12,
    "Sprite":8
}

fig,ax = plt.subplots()
ax.bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax.set(title = "Bach Hoa Xanh's Soft drink Prices",ylabel="Prices ($)")
```

```python
fig,ax=plt.subplots()
ax.barh(soft_drink_prices.keys(),soft_drink_prices.values())
```

# 4.Histogram

```python
# Prepare data
np.random.seed(42)
student_height = np.random.normal(170,10,250)
```

```python
student_height
```

```python
#pylot API
plt.hist(student_height,bins=30);
```

### 5.subplots

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot data to each axis
ax1.plot(x, x/2)
ax1.plot(x, x*2)

ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax4.hist(student_height,bins=30)

plt.show()
```

```python
# Option 2:
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# Plot data to each axis
ax[0,0].plot(x, x/2)
ax[0,0].plot(x, x*2)

ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax[1,1].hist(student_height,bins=30)

plt.show()
```
